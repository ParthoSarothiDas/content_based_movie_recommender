```python
# New Dataframe from movie_df
movie_df = movie_df[['id','title', 'runtime','release_date','overview', 'genres', 'keywords',  'popularity', 'vote_average',
       'vote_count']]
```

```python
movie_df.head()
```

```python
credits_df.head()
```

```python
credits_df = credits_df[['title','cast','crew']]
```

```python
credits_df.head()
```

```python
movie_df = movie_df.merge(credits_df, how='left', on='title')
```

```python
extra_df.head()
```

```python
extra_df.drop(columns='Unnamed: 0', inplace=True)
```

```python
movie_df = movie_df.merge(extra_df, how='left', on='id')
movie_df.head()
```
